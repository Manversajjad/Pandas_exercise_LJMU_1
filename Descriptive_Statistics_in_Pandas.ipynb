```python
import numpy as np
import pandas as pd
```

```python
ser1= pd.Series(np.random.randn(80))
ser1
```

```python
# Descriptive Statistics for a series
ser1.describe()
```

```python
# If wanted to extract particular element from the above, count,mean, std etc. The syntax can be used as follows
ser1.describe()['mean']
```

Generating Descriptive Statistics for a DataFrame

```python
df1 = pd.DataFrame(np.random.rand(3,3), columns=['A', 'B', 'C'], index=[1, 2, 3])
df1
```

```python
df1.describe()
```

```python
# For Dataframe we can use LOC function to iterate
df1.describe().loc['mean']
```

Compute Summary Statistics

```python
df2 = pd.DataFrame(np.random.rand(3,3), columns=['A','B', 'C'], index=[1, 2, 3])
df2
```

```python
# Generate Sum of values for columns
genr = df2.sum(axis=0)
genr
```

```python
# Generate Sum of values for Rows
genr1 = df2.sum(axis=0)
genr1
```

**Practice Problem :- Descriptive Statistics**

```python
product_sales = pd.DataFrame({'Monday': [32, 152, 94, 0], 'Tuesday': [45, 109, 114, 0], \
                             'Wednesday': [15, 77, 85, 0], 'Thursday': [29, 85, 132, 0], \
                             'Friday': [63, 143, 101, 15], 'Saturday': [87, 211, 186, 25]}, \
                             index=['Cupcakes', 'Pie Slices', 'Brownies', 'Cookies'])
product_sales
```

```python
# Generate descriptive statistics for the entire DataFrame.
descriptive_statistics1 = product_sales.describe()
descriptive_statistics1
```

```python
# What is the total number of products sold on each day?
product_sales.sum()
```

```python
# Total number of products sold during the week.
product_sales.sum(axis=1)
```

```python
# For the product type that sold the most across the week, how many were sold?
product_sales.sum(axis=1).max()
```

```python
# For the product type that sold the most across the week, list the name of the product and the number sold.
product_sales.sum(axis=1)[product_sales.sum(axis=1).max() == product_sales.sum(axis=1) ]
```
