```python
import numpy as np
import pandas as pd
```

# Row Filtering

```python
New = pd.DataFrame(np.random.rand(3,3),columns=['a','b','c'], index=[1,2,3])
New
```

```python
# Filtering with the help of Boolean Values
filter = New.loc[np.array([True,False,True])]
filter
```

# Column Filtering

```python
New1 = pd.DataFrame(np.random.rand(3,4),columns=['a','b','c','d'], index=[1,2,3])
New1
```

```python
shuffle =New1.loc[:, np.array([True,False,True,True])]
shuffle
```

# Filtering By Comparision Mask

```python
mask = pd.DataFrame(np.random.rand(3,4),columns=['a','b','c','d'], index=[1,2,3])
mask
```

```python
mask1 = mask > 0.79
mask1
```

```python
# Instead of seeing Boolean values, we can see values which are above 0.79
mask[mask1]
```

```python
mask1
```

```python
# Check all of the columns. For all of the columns in given row, does the row contains atleast one value that meets the Mask variable condition
mask1.any(axis=1)
```

```python
# Now check for the Columns
mask1.any(axis=0)
```
