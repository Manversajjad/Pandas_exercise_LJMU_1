Suppose you are working as a data analyst in the Human Resources Department of an e-commerce company. The senior management wants to analyze the employees who joined the company in February 2022 as part of an ongoing analysis. You have the following information available for each employee:

*  name
*  age
*  gender
*  start_date (formatted as mm/dd/yyyy)
*  salary (monthly salary in US dollars)


| index | name | age | gender | start_date | salary |
| :- | -: | -: | -: | -: | :-: |
| 1 | George | 25 | M | 02-01-2022 | 50000 |
| 2 | Jill | 24 | F | 02-01-2022 | 48000 |
| 3 | Thomas | 24 | M | 02-02-2022 | 45000 |
| 4 | Jim | 27 | F | 02-03-2022 | 60000 |
| 5 | Shelby | 26 | F | 02-04-2022 | 55000 |

Use the following code to create a DataFrame **new_employees** from the data provided to you.

```python
# Importing the required library
import pandas as pd

# Creating a DataFrame new_employees
new_employees = pd.DataFrame({'name': ['George', 'Jill', 'Thomas', 'Jim', 'Shelby'], 'age': [25, 24, 24, 27, 26],
                          'gender': ['M', 'F', 'M', 'F', 'F'], 'start_date': ['02-01-2022', '02-01-2022', '02-02-2022', '02-03-2022', '02-04-2022'],
                          'salary': [50000, 48000, 45000, 60000, 55000]},
                          index=[1, 2, 3, 4, 5])

# Printing the DataFrame
new_employees
```

**Question 1:** How many employees joined the company on February 1, 2022?


A. 1 \
B. 2 \
C. 3 \
D. 4


```python
joined = new_employees['start_date'] == '02-01-2022'
len(new_employees[joined])
```

**Question 2:** Write Python code that outputs the names of the employees whose monthly salaries are less than $50,000.

```python
# Write your code here
new_employees[new_employees['salary'] < 50000]['name']
```
