```python
import numpy as np
import pandas as pd
```

```python
new_data = pd.DataFrame(np.random.rand(4,4), columns=['A','B','C','D'], index=[1,2,3,4])
new_data
```

```python
# changing values using loc method
new_data.loc[1,'A'] = np.random.rand()
new_data
```

```python
new_data['E'] = np.random.rand(4)
new_data
```

```python
new_data['F'] = pd.Series(np.random.rand(4), index=[1,2,3,5])
new_data
```

```python
new_data['F'] = pd.Series(np.random.rand(4), index=[1,2,3,4])
new_data
```

```python
# Dropping the row using Drop() function
deleted = new_data.drop(2, axis=0)
deleted
```
