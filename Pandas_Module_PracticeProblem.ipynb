```python
# import packages
import pandas as pd
import numpy as np
```

```python
# Product_sales DataFrame values
product_sales = pd.DataFrame({'Monday': [32, 152, 94, 0], 'Tuesday': [45, 109, 114, 0], \
                             'Wednesday': [15, 77, 85, 0], 'Thursday': [29, 85, 132, 0], \
                             'Friday': [63, 143, 101, 15], 'Saturday': [87, 211, 186, 25]}, \
                             index=['Cupcakes', 'Pie Slices', 'Brownies', 'Cookies'])
product_sales
```

Note :- If wanted to filter the rows, we can filter only with the help of loc function in pandas

```python
# What were the sales for all days for cupcakes and cookies?
cupCook = product_sales.loc[['Cupcakes', 'Cookies']]
cupCook
```

```python
# What were the sales for all products on Monday, Wednesday, and Friday?
all_sales = product_sales[['Monday', 'Wednesday', 'Friday']]
all_sales
```

```python
# What were the sales for all days for products which had no sales on at least one day?
# no_sale_products = product_sales[(product_sales == 0).axis =1]
no_sales_products = product_sales[(product_sales == 0).any(axis=1)]
no_sales_products
```
